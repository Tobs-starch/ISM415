### QUESTION:  build a machine learning model to predict the 'SALE PRICE', using ideas from the data science workflow, and ML workflow.


```python
import pandas as pd #import library
```

```python
data = pd.read_csv('NYC_Property_Data.csv') #IMPORT DATA
```

```python
data
```

### Data Cleaning

```python
data.isnull().sum() 
```

```python
data.dropna(how = 'any', inplace = True)
```

```python
data.isnull().sum()
```

### REMOVAL OF IRRELEVANT COLUMNS

```python
nyc = data.drop(['Unnamed: 0', 'Unnamed: 0.1','ADDRESS','SALE DATE','SALE PRICE'], axis = 1)
nyc
```

```python
target = data['SALE PRICE']
```

```python
target.head()
```

```python
nyc.head()
```

### IMPORT LABEL ENCODER

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le =  LabelEncoder()
```

### To encode your categorical variable and transforms it into numerical data

```python
nyc['NEIGHBORHOOD'] = le.fit_transform(nyc['NEIGHBORHOOD'])
nyc['BUILDING CLASS CATEGORY'] = le.fit_transform(nyc['BUILDING CLASS CATEGORY'])
nyc['TAX CLASS AT PRESENT'] = le.fit_transform(nyc['TAX CLASS AT PRESENT'])
nyc['BUILDING CLASS AT PRESENT'] = le.fit_transform(nyc['BUILDING CLASS AT PRESENT'])
nyc['BUILDING CLASS AT TIME OF SALE'] = le.fit_transform(nyc['BUILDING CLASS AT TIME OF SALE'])
```

```python
nyc
```

```python
from sklearn.model_selection import train_test_split
```

```python
nyc_train, nyc_test, target_train, target_test = train_test_split(nyc,target, test_size = 0.2)
```

```python
nyc_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
model = regressor.fit(nyc_train, target_train)
```

```python
target_pred = model.predict(nyc_test)
```

## ACTUAL VS PREDICTED DATAFRAME WITH INCLUSION OF ERROR

```python
pd.DataFrame({'Actual': target_test, 'Predicted': target_pred, 'Error': target_test-target_pred})
```
